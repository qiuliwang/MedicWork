```python
'''
Created by WangQL
1.20.2018

将LIDC-IDRI下所有包含CT图像的文件夹（包含文件）复制到另外一个目录
'''

import os
import pandas as pd


# # data_dir = 'stage1'
# # patients = os.listdir(data_dir)
labels_data1 = pd.read_csv('LIDC-IDRI_MetaData.csv',index_col = 0)
labels_data2 = pd.read_csv('LIDC-IDRI_MetaData.csv',index_col = 'Series UID')

print(labels_data1.columns)
print(labels_data2.columns)
```

```python
patients_id = labels_data1.iloc[:,8]
arrs=patients_id.values  
print(type(arrs[0]))
print(arrs[:10])
print(len(arrs))
```

```python
new_patients = []
new_patients_id = []

# get id which has CT info
for patient in arrs:
    #  label = labels_df.at[patient, 'cancer']
    temp_type = labels_data2.at[patient, 'Modality']
    if temp_type == 'CT':
        new_patients.append(patient)
        temp_id = labels_data2.at[patient, 'Patient Id']
        new_patients_id.append(temp_id)
        
print(new_patients[:10])
print(len(new_patients))
print(len(new_patients_id))
```

```python
# combine id with series
dictionary = {}
i = 0

while i < len(new_patients_id):
    id = new_patients_id[i]
    series = new_patients[i]
    if i < 10:
        print(id, series)
    dictionary[id] = series
    i += 1
# import csv

# # with open('test.csv', "w") as f:
# #     writer = csv.writer(f)
# #     for row in new_patients:
# #         row = str(row)
# #         writer.writerow(row)

# # write
# f = open('test.txt', 'w')
# for row in new_patients:
#     f.write(row + "\n")
# f.close()

# # read
# f = open('test.txt', 'r')
# list1 = f.readlines()
# print(len(list1))
# f.close()
```

```python
import shutil

# copy dicom file to new dir

lidc_samples = os.listdir('DOI')
print(lidc_samples[:10])
old_dir = 'DOI\\'
new_dir = 'CT_DATA\\'
i = 0

count = len(lidc_samples)
print('copying')

for lidc_sample in lidc_samples:
    if i % 5 == 0:
        print(i / count)
    i += 1
    templist = os.listdir(old_dir + lidc_sample)
    series = dictionary[lidc_sample]
    for temp in templist:
        #print(temp)
        xxx = os.listdir(old_dir + lidc_sample + '\\' + temp)
#         for x in xxx:
#             print(x)
        if series in xxx:
            pathstr = old_dir + lidc_sample + '\\' + temp + '\\' + series
#             f2.write(pathstr + '\n')
            dicomfiles = os.listdir(pathstr)
            #print(len(dicomfiles))
            
            des_dir = new_dir + series
            os.makedirs(des_dir,0o777,True)
            
            count = len(lidc_samples)
            
            for dicomfile in dicomfiles:
                shutil.copy(pathstr+'\\'+dicomfile, des_dir)

# f2.close()
```
